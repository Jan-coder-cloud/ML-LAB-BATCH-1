```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, scale, Binarizer

# Load the dataset
filename = list(file.keys())[0]
df=pd.read_csv(filename)

# Display original data
print("Original Dataset:\n", df)

# ------------------------------------------------
# 1️⃣ Label Encoding for 'gender' and 'selected'
# ------------------------------------------------
le_gender = LabelEncoder()
df['gender'] = le_gender.fit_transform(df['gender'])  # Female=0, Male=1

le_selected = LabelEncoder()
df['selected'] = le_selected.fit_transform(df['selected'])  # FALSE=0, TRUE=1

print("\nAfter Label Encoding:\n", df)

# ------------------------------------------------
# 2️⃣ Scaling the 'Marks' column
# ------------------------------------------------
df['Scaled_Marks'] = scale(df['Marks'])

print("\nAfter Scaling 'Marks':\n", df[['Marks', 'Scaled_Marks']])

# ------------------------------------------------
# 3️⃣ Binarizer: Convert Marks into Pass/Fail (e.g., >= 50 is Pass)
# ------------------------------------------------
binarizer = Binarizer(threshold=50)
df['Binarized_Marks'] = binarizer.fit_transform(df[['Marks']])

print("\nAfter Binarization of Marks:\n", df[['Marks', 'Binarized_Marks']])
```
